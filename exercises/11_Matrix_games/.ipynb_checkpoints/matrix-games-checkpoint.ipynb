# adaptive EG
The initial question was whether EG is averaged. Maybe we can make it averaged by a particular **adaptive** stepsize choice..

## The Algorithm

```python
def EG(x0, lr, F, max_iter=100, proj=None, adaptive='False', mu=0.5):
    
    if proj is None:
        proj = lambda x: x
    lr1, lr2 = lr
    lr1_old = lr1
    v = x0
    store_iter, store_lr = [x0], np.empty((2, max_iter))

    for k in range(max_iter):
        Fv = F(v)
        # extrapolation step
        u_new = proj(v - lr1 * Fv)
        Fu_new = F(u_new)
        
        #compute stepsize
        if k > 0:
            if adaptive[-2:] == 'BB':
                u_bar = proj(v - lr1_old * Fv)
                diff_grad_norm = np.linalg.norm(Fu_new-Fu)
                if diff_grad_norm < 1e-14:
                    return store_iter, store_lr[:, :k]
                prod = (F(u_bar)-Fu) @ (v-v_old) / diff_grad_norm**2

                lr2 = 2 * mu * prod
                if adaptive[:2] == 'BB':
                    lr1 = lr2
            if adaptive[-3:] == 'Lip':
                lr2 = 0.5 * np.linalg.norm(u_new-v) / np.linalg.norm(Fu_new-Fv)
        
        # store old variables
        store_lr[:, k] = np.array([lr1, lr2])
        v_old, Fv_old, Fu, u = v.copy(), Fv.copy(), Fu_new.copy(), u_new.copy()
        store_iter.append(u.copy())
        
        # update iterate
        v = proj(v - lr2 * Fu_new)
        
        if adaptive[:3] == 'Lip':
            diff_grad_norm = np.linalg.norm(Fu_new-F(v))
            if diff_grad_norm < 1e-14:
                return store_iter, store_lr[:, :k]
            lr1_old = lr1
            lr1 = 0.5 * np.linalg.norm(u_new-v) / diff_grad_norm

    return store_iter, store_lr
```

## bilinear min-max

$$
  \min_x \max_y y^T A x
$$

```python
def make_F(A):
    """saddle gradient of bilinear (square) function"""
    d = A.shape[1]

    def F(z):
        ret = np.zeros(2*d)
        ret[0:d] = A @ z[d:]
        ret[d:] = -A.T @ z[0:d]
        return ret
    return F
```

### 1-dimensional

```python
d = 1
A = np.eye(d)
x0 = np.ones(2*d)
max_iter = 100

F = make_F(A)
```

#### Run methods:

```python
lr = 0.8
xs, stepsize = EG(x0, [lr, lr], F, max_iter)
print(f'*EG*: distance to solution of last itarate {np.linalg.norm(xs[-1])}')
```

```python
xs_adap, stepsizes_adap = EG(x0, [lr, lr], F, max_iter, None, adaptive='Fixed-BB', mu=0.5)
print(f'*adaptive EG*: distance to solution of last itarate {np.linalg.norm(xs_adap[-1])}')
```

#### plotting

```python
cmap = plt.get_cmap('tab10')
cols = cmap(np.linspace(0, 1, 5))
markers = ['o', 'd', 's', 'X']
```

```python
plt_x, plt_y, norms = transform_to_plottable_2d(xs)
plt_x_adap, plt_y_adap, norms_adap = transform_to_plottable_2d(xs_adap)
```

```python
fig, ax = plt.subplots(1, 2)
ax[0].plot(plt_x, plt_y, label=f"EG, lr={lr}", marker='o', markevery=(0.99, 0.99), color=cols[0])
ax[0].plot(plt_x_adap, plt_y_adap, label=f"semi-adaptive EG, lr={lr}", marker='s', markevery=(1.2,0.5), color=cols[1])
ax[1].set_title("iterates")
ax[0].legend()
ax[1].plot(norms, label=f"EG, lr={lr}", marker='o', markevery=(0.1, 0.5), color=cols[0])
ax[1].plot(norms_adap, label=f"semi-adaptive EG, lr={lr}", marker='s', markevery=(0.1,0.5), color=cols[1])
ax[1].set_yscale("log")
ax[1].set_title("distance to solution")
ax[1].legend()
```

### more dimensional

```python
d = 10
np.random.seed(1)
A = np.random.random((d, d))
x0 = np.ones(2*d)
max_iter = 5000

F = make_F(A)
```

Compute Lipschitz constant for optimal fixed stepsize.

```python
lambda_max = np.linalg.norm(A, 2)
print(f'largest eigenvalue of A is {lambda_max}')
lr = 0.9/lambda_max
```

All the stepsize rules to try:

```python
modes = ('No', 'Fixed-BB', 'Lip-BB', 'Lip-Lip')
```

#### Run experiments

```python
plt_X, stepsizes, norms = [], [], []
for adaptive in modes:
    xs, steps = EG(x0, [lr, lr], F, max_iter, adaptive=adaptive)
    XX, norm, _ = transform_to_plottable(xs)
    plt_X.append(XX)
    norms.append(norm)
    stepsizes.append(steps)
```

```python
PCA_plot(plt_X, modes, cols)
```

```python
stepsize_plot(cols, stepsizes, modes, markers)
```

```python
fig, ax = plt.subplots()
for norm, mode, mark, col in zip(norms, modes, markers, cols):
    plt.plot(range(max_iter+1), norm, label=f"EG, {mode}, lr={lr:.2f}", 
             marker=mark, markevery=0.1 ,alpha=0.6, linewidth=2, markersize=10, color=col)
plt.legend()
plt.title("distance to solution")
plt.yscale('log')
plt.xscale('log')
```

# Policeman and Burgler
Bilinear problem with constraints

```python
np.random.seed(44)
from problem_instances import policeman_and_burglar_matrix
```

```python
d = 500
max_iter = 1000
A = policeman_and_burglar_matrix(d)
sp_norm = np.linalg.norm(A, 2)
def gap_fn(iterate):
    return np.max(A.T @ iterate[:d]) - np.min(A @ iterate[d:])
def proj(z):
    return np.concatenate((projection_simplex(z[:d]), projection_simplex(z[d:])))
F_burgler = make_F(A)
x0 = np.ones(2*d)/(d)
```

The larger we choose the stepsize the better

```python
lr = 0.99/sp_norm
```

```python
plt_X, stepsizes, norms, gaps = [], [], [], []
for adaptive in modes:
    xs, steps = EG(x0, [lr, lr], F_burgler, max_iter, proj, adaptive=adaptive)
    XX, norm, gap = transform_to_plottable(xs, gap_fn)
    plt_X.append(XX)
    norms.append(norm)
    stepsizes.append(steps)
    gaps.append(gap)
```

do the plotting

```python
PCA_plot(plt_X, modes, cols)
```

```python
fig, ax = plt.subplots()
for gap, mode, mark, col in zip(gaps, modes, markers, cols): 
    plt.plot(gap, label=f"EG, {mode}, lr={lr:.2f}", 
             #marker=mark, markevery=0.1 ,alpha=0.6, linewidth=2, markersize=10, 
             color=col)
plt.legend()
plt.yscale("log")
#plt.xscale("log")
plt.title("primal dual gap")
```
